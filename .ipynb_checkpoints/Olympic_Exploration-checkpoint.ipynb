```python
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None
import matplotlib.pyplot as plt
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC

from DataProcTools.DataCleaning.MissingValues import get_nan_values_percentage
```

```python
dataset = pd.read_csv('athlete_events.csv')

# Initial dataset info before imputation
dataset.info()
dataset.head()
```

```python
print(dataset.describe())
missingValuesInfo = get_nan_values_percentage(dataset)
print()
print(missingValuesInfo)
```

```python
# Average age per NOC ( country ) for every team entry
dataset[['Age','NOC']].groupby('NOC').Age.apply(lambda x: x.mode())
```

```python
# Get average height and weight based & groupby NOC
average_height = dataset[['Height','NOC']].groupby('NOC').mean()
average_weight = dataset[['Weight','NOC']].groupby('NOC').mean()

# Get average age & store as float64
average_age = dataset['Age'].mode().loc[0]
```

```python
# Defining imputing functions
def height_imputer(dataset, average_height):
    for row in dataset.index:    
        if((dataset['Height'].isna()[row])):        
            NOC_value = dataset['NOC'][row]
            height_per_NOC = average_height.loc[NOC_value, 'Height']
            dataset['Height'].fillna(height_per_NOC, inplace=True)
            
def weight_imputer(dataset, average_weight):
    for row in dataset.index:
        if((dataset['Weight'].isna()[row])):
            NOC_value = dataset['NOC'][row]
            weight_per_NOC = average_weight.loc[NOC_value, 'Weight']
            dataset['Weight'].fillna(weight_per_NOC, inplace=True)    

def age_imputer(dataset, average_age): 
    dataset['Age'].fillna(average_age, inplace=True)

def medal_imputer(dataset):
    dataset['Medal'].fillna('No_Medal', inplace=True)

# Calling our functions
age_imputer(dataset, average_age)
height_imputer(dataset, average_height)
weight_imputer(dataset, average_weight)
medal_imputer(dataset)
```

```python
# Question 1: Relationship between obesity rates in participants and NOC ( country )
# 1. Get BMI ( weight/ height^2 )
# 2. Calculate Body Fat Percentage (1.20 x BMI) + (0.23 x Age) - 16.2 [ if female = 5.4 ] = Body Fat Percentage.

avg_shwn_df = pd.DataFrame(dataset[['NOC','Sex','Height','Weight', 'Age']], columns = ['NOC', 'Sex', 'Height', 'Weight', 'Age'])

# Function to return dataset with only males / females based on input
def get_gender_df(dataset, sex):
    gender_df = avg_shwn_df[avg_shwn_df['Sex'] == sex]
    return gender_df

males_df = get_gender_df(dataset, "M")
females_df = get_gender_df(dataset, "F")

# calculating BMI for males / females
males_df['BMI'] = ( avg_shwn_df['Weight'] / avg_shwn_df['Height']**2 ) * 10000
females_df['BMI'] = ( avg_shwn_df['Weight'] / avg_shwn_df['Height']**2 ) * 10000

# calculating body fat percentage for males / females
males_df['AvgBF%'] = ( (1.20 * males_df['BMI']) + (0.23 * males_df['Age']) - 16.2)
females_df['AvgBF%'] = ( (1.20 * females_df['BMI']) + (0.23 * females_df['Age']) - 5.4)

print("Male Average")
display(males_df.groupby('NOC').mean().sort_values('AvgBF%',ascending=False))
print("Female Average")
display(females_df.groupby('NOC').mean().sort_values('AvgBF%',ascending=False))
```
